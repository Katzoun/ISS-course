```python
# -------------- Cviko 1  --------------------------
import numpy as np
import matplotlib.pyplot as plt
N = 8
n = np.arange(N)
x = [1, 1, 1, 0, 0, 0.5, 0.5, 0]
c = np.zeros(5)
for k in [0,1,2,3,4]: 
    a = np.cos(2 * np.pi * k / N *n)
    print("----------", k)
    print("a:",a)
    print("ax:",a*x)
    c[k] = np.sum(a*x) 
    print(c[k])
    
```

```python
d = np.zeros(5)
for k in [0,1,2,3,4]: 
    b = np.sin(2 * np.pi * k / N *n)
    print("----------", k)
    print("a:",b)
    print("ax:",b*x)
    d[k] = np.sum(b*x) 
    print(d[k])
    
```

```python
X = c + 1j*d
print(np.abs(X))
print(np.angle(X))

Xfft = np.fft.fft(x)
print(np.abs(Xfft[0:5]))
print(np.angle(Xfft[0:5]))
```

```python
# synthesizing them ... 
xs = np.zeros(N)
k=0
this = 1/N*np.abs(Xfft[k]) * np.cos(2 * np.pi * k / N *n + np.angle(Xfft[k]))
xs = xs + this 
for k in [1,2,3]:
    this = 1/N*2 * np.abs(Xfft[k]) * np.cos(2 * np.pi * k / N *n + np.angle(Xfft[k]))
    xs = xs + this 
k=4
this = 1/N*np.abs(Xfft[k]) * np.cos(2 * np.pi * k / N *n + np.angle(Xfft[k]))
xs = xs + this 
print(xs)
```
