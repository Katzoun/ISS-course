```python
# -------------- Math for ISS --------------------------
import numpy as np
import matplotlib.pyplot as plt

# >>>>>>>>>>>>>>>>> #basicops <<<<<<<<<<<<<<<<<<<<<<<<<
# --- addition of signals ---
t = np.arange(-5,5,0.01)
x1 = np.zeros(t.shape)
x1[(t>-1) & (t<0.5)] = 0.5
x2 = np.zeros(t.shape)
x2[(t>-0.5) & (t<2)] = 1
y = x1 + x2
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].plot(t,x1)
ax[0].set_ylabel('$x_1(t)$')
ax[0].set_ylim(-0.1,2)
ax[0].grid()
ax[1].plot(t,x2,color='g')
ax[1].set_ylabel('$x_2(t)$')
ax[1].set_ylim(-0.1,2)
ax[1].grid()
ax[2].plot(t,y,color='r',linewidth=3)
ax[2].set_ylabel('$y(t)$')
ax[2].set_xlabel('$t$')
ax[2].set_ylim(-0.1,2)
ax[2].grid()
#plt.savefig('FIG/soucet.pdf')
```

```python
# --- difference of signals ---
t = np.arange(-5,5,0.01)
x1 = np.zeros(t.shape)
x1[(t>-1) & (t<0.5)] = 0.5
x2 = np.zeros(t.shape)
x2[(t>-0.5) & (t<2)] = 1
y = x1 - x2
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].plot(t,x1)
ax[0].set_ylabel('$x_1(t)$')
ax[0].set_ylim(-1.1,1.1)
ax[0].grid()
ax[1].plot(t,x2,color='g')
ax[1].set_ylabel('$x_2(t)$')
ax[1].set_ylim(-1.1,1.1)
ax[1].grid()
ax[2].plot(t,y,color='r',linewidth=3)
ax[2].set_ylabel('$y(t)$')
ax[2].set_xlabel('$t$')
ax[2].set_ylim(-1.1,1.1)
ax[2].grid()
#plt.savefig('FIG/rozdil.pdf')
```

```python
# --- product signalu ---
t = np.arange(-5,5,0.01)
x1 = np.zeros(t.shape)
x1[(t>-1) & (t<0.5)] = 0.5
x2 = np.zeros(t.shape)
x2[(t>-0.5) & (t<2)] = 1
y = x1 * x2
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].plot(t,x1)
ax[0].set_ylabel('$x_1(t)$')
ax[0].set_ylim(-0.1,1.1)
ax[0].grid()
ax[1].plot(t,x2,color='g')
ax[1].set_ylabel('$x_2(t)$')
ax[1].set_ylim(-0.1,1.1)
ax[1].grid()
ax[2].plot(t,y,color='r',linewidth=3)
ax[2].set_ylabel('$y(t)$')
ax[2].set_xlabel('$t$')
ax[2].set_ylim(-0.1,1.1)
ax[2].grid()
#plt.savefig('FIG/soucin.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #basicops_seq <<<<<<<<<<<<<<<<<<<<<<<<<
n = np.arange(-10,10)
x1 = n / 10
x2 = np.zeros(n.shape)
x2[5:] = 2
_, ax = plt.subplots(5,1, figsize=(10,7))
ax[0].stem(n,x1,basefmt=" ")  # don't want to have baseline...
ax[0].set_ylabel('$x_1[n]$')
ax[0].set_ylim(-1.1,2.1)
ax[0].set_xticks([-10, -5, 0, 5, 10])
ax[0].grid()
ax[1].stem(n,x2,'g',basefmt=" ",markerfmt='go')  # don't want to have baseline...
ax[1].set_ylabel('$x_2[n]$')
ax[1].set_ylim(-1.1,2.1)
ax[1].set_xticks([-10, -5, 0, 5, 10])
ax[1].grid()
y = x1 + x2
ax[2].stem(n,y,'r',basefmt=" ",markerfmt='ro')  # don't want to have baseline...
ax[2].set_ylabel('$x_1[n] + x_2[n]$')
ax[2].set_xticks([-10, -5, 0, 5, 10])
ax[2].grid()
y = x2 - x1
ax[3].stem(n,y,'r',basefmt=" ",markerfmt='ro')  # don't want to have baseline...
ax[3].set_ylabel('$x_2[n] - x_1[n]$')
ax[3].set_xticks([-10, -5, 0, 5, 10])
ax[3].grid()
y = x2 * x1
ax[4].stem(n,y,'r',basefmt=" ",markerfmt='ro')  # don't want to have baseline...
ax[4].set_ylabel('$x_1[n] x_2[n]$')
ax[4].set_xticks([-10, -5, 0, 5, 10])
ax[4].grid()
ax[4].set_xlabel('$n$')
#plt.savefig('FIG/diskretni_all_ops.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #linear <<<<<<<<<<<<<<<<<<<<<<<<<
t = np.arange(-5,5,0.01)
x1 = 3 * t;
x2 = 2*t + 1;
x3 = -t;
x4 = -2*t -2;
plt.figure(figsize=(10,5))
plt.plot(t,x1,label="$x_1(t) = 3t$")
plt.plot(t,x2,label="$x_2(t) = 2t+1$")
plt.plot(t,x3,label="$x_3(t) = -t$")
plt.plot(t,x4,label="$x_4(t) = -2t-2$")
plt.gca().set_xlabel('$t$')
plt.gca().set_ylabel('$x(t)$')
plt.gca().grid()
plt.legend()
#plt.savefig('FIG/lin.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #deriv_int_anal <<<<<<<<<<<<<<<<<<<<<<<<<
Delta = 0.05
t = np.arange(-3,2,Delta)
def f_direct(t):
    return np.power(t,2) + 2*t + 3
def f_deriv(t):
    return 2*t + 2
def f_primitive(t):
    return np.power(t,3) / 3 + np.power(t,2) + 3*t

x = f_direct(t)
xd = f_deriv(t)
xp = f_primitive(t)

plt.figure(figsize=(8,6))
plt.plot(t,x,label="$x(t) = t^2 + 2t + 3$")
plt.plot(t,xd,label="derivation")
plt.plot(t,xp,label="prim. function")
plt.gca().set_xlabel('$t$')
plt.gca().grid()
plt.legend()
#plt.savefig('FIG/polynom_derivace_prim.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #deriv_num <<<<<<<<<<<<<<<<<<<<<<<<<
# for t1 = 0
t1 = 0
t2 = t1 + Delta
presne = f_deriv(t1)
priblizne = (f_direct(t2) - f_direct(t1)) / (t2 - t1)
print(presne,priblizne)
```

```python
# for t1 = 1
t1 = 1
t2 = t1 + Delta
presne = f_deriv(t1)
priblizne = (f_direct(t2) - f_direct(t1)) / (t2 - t1)
print(presne,priblizne)
```

```python
# drawing tangents to prove that we're OK ... derive the formula if you want ;)
plt.figure(figsize=(8,6))
t1 = 0
tecna1 = f_deriv(t1) * t + f_direct(t1) - f_deriv(t1) * t1
t1 = 1
tecna2 = f_deriv(t1) * t + f_direct(t1) - f_deriv(t1) * t1
plt.plot(t,x,label="$x(t) = t^2 + 2t + 3$")
plt.plot(t,tecna1,label="tecna pro $t = 0$")
plt.plot(t,tecna2,label="tecna pro $t = 1$")
plt.gca().set_xlabel('$t$')
plt.gca().set_xlim(-1,1.5) # zoom needed
plt.gca().set_ylim(-1,8)   # zoom needed
plt.gca().grid()
plt.legend()
#plt.savefig('FIG/tecny.pdf')
```

```python
# >>>>>>>>>>>>>>>>> #int_anal_num <<<<<<<<<<<<<<<<<<<<<<<<<
t1 = 0
t2 = 1

# exactly using primitive function
int_presne = f_primitive(t2) - f_primitive(t1)

# numerically - mind how to best find the incides for t1 and t2 !
ii1 = np.argmin(np.abs(t - t1)) # just to be sure, t1 might NOT be any of the generated times !
ii2 = np.argmin(np.abs(t - t2)) # dtto
int_numericky = np.sum(x[ii1:ii2] * Delta)
print(int_presne, int_numericky)

# vizualization
tvisu = np.arange (0,1,Delta)
xvisu = f_direct(tvisu)
plt.figure(figsize=(10,4))
plt.bar(tvisu,xvisu,Delta,linewidth=1,align='edge',edgecolor='black')
plt.gca().set_xlabel('$t$')
plt.gca().set_ylabel('$x(t)$')
plt.gca().grid()
#plt.savefig('FIG/numint.pdf')
```
